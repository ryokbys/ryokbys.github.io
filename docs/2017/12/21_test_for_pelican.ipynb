グラフもいけるらしい．

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style=('ticks')
sns.set_context('talk')

x = np.linspace(0.,4.*np.pi,200)
y = np.sin(x)
plt.plot(x,y,'r-',label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```
